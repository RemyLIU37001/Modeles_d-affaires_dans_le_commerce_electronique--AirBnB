### 4. La visualisation des données 

#### 4.1 La correlation entre les variables 

```python
corr = data.corr(method='kendall')
plt.figure(figsize=(15,8))
sns.heatmap(corr, annot=True)
data.columns
```

#### 4.2 La visualisation par le pricipales variables 

#### 4.2.1 Neighbourhood Group

```python
sns.countplot(data['neighbourhood_group'], palette="plasma")
fig = plt.gcf()
fig.set_size_inches(10,10)
plt.title('Neighbourhood Group')
```

#### 4.2.2 Neighbourhood 

```python
sns.countplot(data['neighbourhood'], palette="plasma")
fig = plt.gcf()
fig.set_size_inches(25,6)
plt.title('Neighbourhood')
```

#### 4.2.3 Room type

```python
#Restaurants delivering Online or not
sns.countplot(data['room_type'], palette="plasma")
fig = plt.gcf()
fig.set_size_inches(10,10)
plt.title('Restaurants livrant en ligne ou non')
```

#### 4.2.4 La relation entre la variablle neighhourhood group et availability_365

```python
plt.figure(figsize=(10,10))
ax = sns.boxplot(data=data, x='neighbourhood_group',y='availability_365',palette='plasma')
```
